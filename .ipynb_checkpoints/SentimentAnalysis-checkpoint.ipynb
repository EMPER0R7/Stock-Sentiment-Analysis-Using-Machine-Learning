# STOCK SENTIMENT ANALYSIS PROJECT

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ta
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold, TimeSeriesSplit
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, roc_curve, roc_auc_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv1D, Flatten, MaxPooling1D
from sklearn.utils import resample
```

```python
class Color:
    PURPLE = '\033[95m'
    CYAN = '\033[96m'
    DARKCYAN = '\033[36m'
    BLUE = '\033[94m'
    GREEN = '\033[92m'
    YELLOW = '\033[93m'
    RED = '\033[91m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'
    END = '\033[0m'
```

```python
news_df = pd.read_csv('final_nyt.csv', parse_dates=['Datetime'])#importing scrapped data
# Convert datetime to date for grouping
news_df['Date'] = news_df['Datetime'].dt.date
news_df['Headline'] = news_df['Headline'].astype(str)
# Group headlines by date
news_df = news_df.groupby('Date')['Headline'].apply(lambda x: ' '.join(x)).reset_index()
```

```python
import warnings
warnings.filterwarnings('ignore', message="Do not pass an `input_shape`/`input_dim` argument to a layer.*")


def analyze_stock(stock_data, news_df):
    # Convert 'Date' columns to datetime
    stock_data['Date'] = pd.to_datetime(stock_data['Date'])
    news_df['Date'] = pd.to_datetime(news_df['Date'])

    # Merge based on 'Date'
    combined_df = pd.merge(stock_data, news_df, on='Date', how='inner')

    # Sentiment Analysis
    analyzer = SentimentIntensityAnalyzer()

    def get_vader_sentiment(text):
        return analyzer.polarity_scores(text)['compound']

    def get_textblob_sentiment(text):
        return TextBlob(text).sentiment.polarity

    combined_df['vader_sentiment'] = combined_df['Headline'].apply(get_vader_sentiment)
    combined_df['textblob_sentiment'] = combined_df['Headline'].apply(get_textblob_sentiment)

    # Add technical indicators
    stock_data['SMA'] = ta.trend.sma_indicator(stock_data['Close'], window=14)
    stock_data['RSI'] = ta.momentum.rsi(stock_data['Close'], window=14)

    combined_df['SMA'] = combined_df['Date'].map(stock_data.set_index('Date')['SMA'])
    combined_df['RSI'] = combined_df['Date'].map(stock_data.set_index('Date')['RSI'])

    # Prepare data for training
    combined_df['return'] = combined_df['Close'].pct_change().shift(-1)
    combined_df['label'] = (combined_df['return'] > 0).astype(int)
    combined_df.dropna(subset=['vader_sentiment', 'textblob_sentiment', 'SMA', 'RSI', 'return'], inplace=True)
    features = ['vader_sentiment', 'textblob_sentiment','SMA', 'RSI']
    X = combined_df[features]
    y = combined_df['label']
    # Address class imbalance
    X_upsampled, y_upsampled = resample(X[y == 1], y[y == 1],
                                        replace=True,
                                        n_samples=X[y == 0].shape[0],
                                        random_state=42)
    X = pd.concat([X[y == 0], X_upsampled])
    y = pd.concat([y[y == 0], y_upsampled])
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    random_seed = 42
   # Initialize models
    models = {
        'Logistic Regression': LogisticRegression(random_state=random_seed),
        'Random Forest': RandomForestClassifier(random_state=random_seed),
        'SVC': SVC(probability=True, random_state=random_seed),
    }

    # Evaluate models using cross-validation
    for name, model in models.items():
        scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')
        print(f"{name} Accuracy: {np.mean(scores):.4f}")

    # Add CNN model
    cnn_model = Sequential()
    cnn_model.add(Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(X_train.shape[1], 1)))
    cnn_model.add(MaxPooling1D(pool_size=2))
    cnn_model.add(Flatten())
    cnn_model.add(Dense(50, activation='relu'))
    cnn_model.add(Dense(1, activation='sigmoid'))
    cnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    # Reshape data for CNN
    X_train_cnn = X_train.values.reshape(X_train.shape[0], X_train.shape[1], 1)
    X_test_cnn = X_test.values.reshape(X_test.shape[0], X_test.shape[1], 1)

    # Train CNN model
    cnn_model.fit(X_train_cnn, y_train, epochs=10, verbose=0)
    cnn_accuracy = cnn_model.evaluate(X_test_cnn, y_test, verbose=0)[1]
    print(f"CNN Accuracy: {cnn_accuracy:.4f}")

    # Select the best model based on accuracy
    best_model_name = max(models, key=lambda name: np.mean(cross_val_score(models[name], X_train, y_train, cv=5, scoring='accuracy')))
    if cnn_accuracy > max(np.mean(cross_val_score(models[name], X_train, y_train, cv=5, scoring='accuracy')) for name in models):
        best_model_name = 'CNN'
        best_model = cnn_model
        X_train_best = X_train_cnn
        X_test_best = X_test_cnn
    else:
        best_model = models[best_model_name]
        best_model.fit(X_train, y_train)
        X_train_best = X_train
        X_test_best = X_test

    print(f"Best model: {best_model_name}")

    if best_model_name == 'CNN':
        y_pred = (best_model.predict(X_test_best) > 0.5).astype("int32")
    else:
        y_pred = best_model.predict(X_test_best)

    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred,zero_division=0)
    print("Classification Report:\n", report)
    print(f"Accuracy: {Color.BLUE}{accuracy:.2%}{Color.END}")

    # Generate trading signals and backtest strategy
    if best_model_name == 'CNN':
        combined_df['signal'] = (best_model.predict(combined_df[features].values.reshape(combined_df[features].shape[0], combined_df[features].shape[1], 1)) > 0.5).astype("int32")
    else:
        combined_df['signal'] = best_model.predict(combined_df[features])

    def backtest_strategy(df, initial_balance=10000):
        balance = initial_balance
        position = 0
        for i, row in df.iterrows():
            if row['signal'] == 1 and position == 0:
                position = balance / row['Close']
                balance = 0
            elif row['signal'] == 0 and position > 0:
                balance = position * row['Close']
                position = 0
        final_balance = balance if position == 0 else position * df.iloc[-1]['Close']
        return final_balance

    final_balance = backtest_strategy(combined_df)

    def calculate_returns(df, initial_investment=10000):
        df['daily_return'] = df['Close'].pct_change()
        df['strategy_return'] = df['signal'].shift(1) * df['daily_return']
        df['cumulative_strategy_return'] = (1 + df['strategy_return']).cumprod()
        df['cumulative_market_return'] = (1 + df['daily_return']).cumprod()

        # Calculate final portfolio value
        final_portfolio_value = initial_investment * df['cumulative_strategy_return'].iloc[-1]
        total_return = final_portfolio_value - initial_investment

        return df, final_portfolio_value, total_return

    combined_df, final_portfolio_value, total_return = calculate_returns(combined_df)
    print(f"{Color.RED}Taking Initial Investment as $10000{Color.END}")
    print(f"Final Portfolio Value:{Color.GREEN} ${final_portfolio_value:.2f}{Color.END}")
    print(f"Total Return:{Color.GREEN} ${total_return:.2f}{Color.END}")

    def calculate_sharpe_ratio(df, risk_free_rate=0.01):
        excess_returns = df['strategy_return'] - risk_free_rate / 252
        sharpe_ratio = np.mean(excess_returns) / np.std(excess_returns) * np.sqrt(252)
        return sharpe_ratio

    sharpe_ratio = calculate_sharpe_ratio(combined_df)
    print(f"Sharpe Ratio: {Color.BLUE}{sharpe_ratio:.2f}{Color.END}")

    def calculate_max_drawdown(df):
        df['cumulative_max'] = df['cumulative_strategy_return'].cummax()
        df['drawdown'] = df['cumulative_strategy_return'] / df['cumulative_max'] - 1
        max_drawdown = df['drawdown'].min()
        return max_drawdown

    max_drawdown = calculate_max_drawdown(combined_df)
    print(f"Maximum Drawdown: {Color.RED}{max_drawdown:.2%}{Color.END}")

    def calculate_trade_stats(df):
        buy_signals = df[df['signal'] == 1]
        sell_signals = df[df['signal'] == 0]

        number_of_trades = len(buy_signals)

        win_trades = (df['strategy_return'] > 0).sum()
        total_trades = len(df[df['signal'].shift(1).notna()])

        win_ratio = win_trades / total_trades

        return number_of_trades, win_ratio

    number_of_trades, win_ratio = calculate_trade_stats(combined_df)
    print(f"Number of Trades Executed: {number_of_trades}")
    print(f"Win Ratio: {Color.BLUE}{win_ratio:.2%}{Color.END}")

    fpr, tpr, thresholds = roc_curve(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred)

    # Plot ROC curve
    plt.figure()
    plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc='lower right')
    plt.show()

    def trading_strategy(data, initial_cash=100000):
        cash = initial_cash
        holdings = 0
        buy_signals = []
        sell_signals = []

        for i in range(len(data)):
            if data['signal'].iloc[i] == 1 and cash >= data['Close'].iloc[i]:
                # Buy one stock
                holdings += 1
                cash -= data['Close'].iloc[i]
                buy_signals.append(i)
            elif data['signal'].iloc[i] == 0 and holdings > 0:
                # Sell one stock
                holdings -= 1
                cash += data['Close'].iloc[i]
                sell_signals.append(i)

        # Calculate final portfolio value
        portfolio_value = cash + holdings * data['Close'].iloc[-1]
        return portfolio_value, buy_signals, sell_signals

   # Filter data for one year (e.g., 2022)
    start_date = '2018-09-01'
    end_date = '2019-09-30'
    filtered_df = combined_df[(combined_df['Date'] >= start_date) & (combined_df['Date'] <= end_date)]

    # Apply the trading strategy to the filtered data
    final_portfolio_value, buy_signals, sell_signals = trading_strategy(filtered_df)

    # Plot the stock price with buy and sell points for the filtered data
    plt.figure(figsize=(16, 8))
    plt.plot(filtered_df['Date'], filtered_df['Close'], label='Close Price', color='blue')
    plt.scatter(filtered_df['Date'].iloc[buy_signals], filtered_df['Close'].iloc[buy_signals], label='Buy Signal', marker='^', color='green', alpha=1)
    plt.scatter(filtered_df['Date'].iloc[sell_signals], filtered_df['Close'].iloc[sell_signals], label='Sell Signal', marker='v', color='red', alpha=1)
    plt.title('Stock Price with Buy and Sell Signals (2022)')
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()
```
